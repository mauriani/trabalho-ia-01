```python
import pandas as pd
import numpy as np
import requests

data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/communities/communities.data'
columns_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/communities/communities.names'

response = requests.get(columns_url)
columns = []


for line in response.text.splitlines():
    # pega os nomes das colunas depois do '@attribute'
    if line.startswith('@attribute'):
        column_name = line.split()[1]
        columns.append(column_name)


ds = pd.read_csv(data_url, header=None, names=columns, na_values='?')

ds.head()
```

```python
def avaliaQualidadeDataframe(df):
  if df.isnull().values.any():
    print('Valores nulos encontrados.')
  else:
    print('Sem valores nulos.')
  for col in df.dtypes.to_list():
    if str(col) not in ["int64", "float64"]:
      print('Coluna com tipo incorreto para análise.')
```

```python
# Checa qualidade
avaliaQualidadeDataframe(ds)
```

```python
# Removendo colunas com valores faltantes
ds_clean = ds.dropna(axis=0)

avaliaQualidadeDataframe(ds_clean)
```

```python
# Removendo colunas NÃO númericas que não contribuiu para a regressão
# exemplo de coluna que está sendo removida é a de communityname

data = ds_clean.select_dtypes(include=['float64', 'int64'])

data.head()
```

```python
# Nós queremos prever a váriavel ViolentCrimesPerPop para o nosso modelo de regressão linear

X = data.drop(columns=['ViolentCrimesPerPop'])
y = data['ViolentCrimesPerPop']

```

```python
# 1.B

# Divisao em treino e teste

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error


# criando o modelo e o treino
model = LinearRegression()
model.fit(X_train, y_train)


# Faz previções

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Agora vamos fazer as métricas RMSE (Root Mean Square Error) e MAE (Mean Absolute Error) que sao
# usadas ara avaliar a precisao de modelos de previsao

rmse_train = mean_squared_error(y_train, y_train_pred, squared=False)
mae_train = mean_absolute_error(y_train, y_train_pred)

rmse_test = mean_squared_error(y_test, y_test_pred, squared=False)
mae_test = mean_absolute_error(y_test, y_test_pred)

print(f'RMSE (Root Mean Square Error): {rmse_train}, MAE: {mae_train}')
print(f'MAE (Mean Absolute Error: {rmse_test}, MAE: {mae_test}')
```

```python
# 1.C

# Aplicando PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

print(f'Variancia - {pca.explained_variance_ratio_}')
```

```python
model_pca = LinearRegression()
model_pca.fit(X_train_pca, y_train)

y_train_pred_pca = model_pca.predict(X_train_pca)
y_test_pred_pca = model_pca.predict(X_test_pca)

# Avaliar o modelo com PCA
rmse_train_pca = mean_squared_error(y_train, y_train_pred_pca, squared=False)
mae_train_pca = mean_absolute_error(y_train, y_train_pred_pca)

rmse_test_pca = mean_squared_error(y_test, y_test_pred_pca, squared=False)
mae_test_pca = mean_absolute_error(y_test, y_test_pred_pca)

print(f'Treino de RMSE COM PCA: {rmse_train}, MAE: {mae_train}')
print(f'Teste de MAE COM PCA: {rmse_test}, MAE: {mae_test}')
```
